```python
import pandas as pd
import json

# Helper function to safely parse the reviews field
def parse_reviews(reviews):
    try:
        # Ensure valid JSON format
        if isinstance(reviews, str):
            return json.loads(reviews.replace("'", '"'))  # Replace single quotes with double quotes
        return []  # Default to empty list if reviews is not a string
    except json.JSONDecodeError:
        return []  # Return empty list for invalid JSON

# Main function to create datasets
def create_datasets(data):
    # Dataset 1: Product-Product Connections
    products = []
    product_edges = []
    for index, row in data.iterrows():
        product_id = row['asin']
        title = row['title']
        group = row['group']
        salesrank = row['salesrank']

        # Add product node details
        similar_products = row['similar'].split() if isinstance(row['similar'], str) else []
        num_similar_products = len(similar_products)
        products.append({
            'asin': product_id,
            'title': title,
            'group': group,
            'salesrank': salesrank,
            'num_similar_products': num_similar_products
        })

        # Add product edges
        for similar_id in similar_products:
            if similar_id.endswith('X'):
                similar_ids_expanded = [f"{similar_id[:-1]}{i}" for i in range(10)]
                product_edges.extend([
                    {'source': product_id, 'target': sim_id, 'relationship': 'similar'}
                    for sim_id in similar_ids_expanded
                ])
            else:
                product_edges.append({'source': product_id, 'target': similar_id, 'relationship': 'similar'})

    # Convert product data to DataFrame
    products_df = pd.DataFrame(products)
    product_edges_df = pd.DataFrame(product_edges)

    # Dataset 2: Customer-Product Connections
    customer_nodes = {}
    customer_edges = []
    for index, row in data.iterrows():
        product_id = row['asin']
        reviews = parse_reviews(row['reviews'])  # Safely parse the reviews field

        # Calculate max helpful and votes for normalization
        max_helpful = max([review['helpful'] for review in reviews], default=0)
        max_votes = max([review['votes'] for review in reviews], default=0)

        for review in reviews:
            customer_id = review['customer_id']
            rating = review['rating']
            helpful = review['helpful']
            votes = review['votes']
            review_date = review['date']

            # Update customer node info
            if customer_id not in customer_nodes:
                customer_nodes[customer_id] = 0
            customer_nodes[customer_id] += 1

            # Calculate weight for the edge
            weight = (
                rating *
                (1 + (helpful + 1) / max(max_helpful + 1, 1)) *
                ((votes + 1) / max(max_votes + 1, 1))
            )

            # Add customer-product edge
            customer_edges.append({
                'customer_id': customer_id,
                'product_id': product_id,
                'weight': weight,
                'timestamp': review_date
            })

    # Convert customer data to DataFrame
    customer_nodes_df = pd.DataFrame([{'customer_id': k, 'num_reviews': v} for k, v in customer_nodes.items()])
    customer_edges_df = pd.DataFrame(customer_edges)

    # Save datasets to CSV
    products_df.to_csv("/content/drive/MyDrive/dataset/SNA/product_nodes.csv", index=False)
    product_edges_df.to_csv("/content/drive/MyDrive/dataset/SNA/product_edges.csv", index=False)
    customer_nodes_df.to_csv("/content/drive/MyDrive/dataset/SNA/customer_nodes.csv", index=False)
    customer_edges_df.to_csv("/content/drive/MyDrive/dataset/SNA/customer_edges.csv", index=False)

    return "Datasets saved successfully."

# Assuming 'data' is already loaded as a DataFrame
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import random
from matplotlib.lines import Line2D

# Set a random seed for reproducibility


# Assuming G is already loaded and nodes have 'node_type' attribute

# Step 1a: Filter nodes by 'node_type' attribute
product_nodes = [node for node, data in G.nodes(data=True) if data.get('node_type', '').lower() == 'product']
customer_nodes = [node for node, data in G.nodes(data=True) if data.get('node_type', '').lower() == 'customer']

# Step 1b: Select Multiple Seed Nodes with Half from 'product' and Half from 'customer'
def select_seed_nodes(G, product_nodes, customer_nodes, num_seeds):
    half_num_seeds = num_seeds // 2
    remaining_seeds = num_seeds - (half_num_seeds * 2)

    selected_product_seeds = []
    selected_customer_seeds = []

    # Select seeds from 'product' nodes
    if len(product_nodes) >= half_num_seeds:
        selected_product_seeds = random.sample(product_nodes, half_num_seeds)
    else:
        selected_product_seeds = product_nodes.copy()
        remaining_seeds += (half_num_seeds - len(selected_product_seeds))

    # Select seeds from 'customer' nodes
    if len(customer_nodes) >= half_num_seeds:
        selected_customer_seeds = random.sample(customer_nodes, half_num_seeds)
    else:
        selected_customer_seeds = customer_nodes.copy()
        remaining_seeds += (half_num_seeds - len(selected_customer_seeds))

    # If there are remaining seeds, select randomly from all nodes excluding already selected
    if remaining_seeds > 0:
        remaining_pool = list(set(G.nodes()) - set(selected_product_seeds) - set(selected_customer_seeds))
        if len(remaining_pool) < remaining_seeds:
            raise ValueError("Not enough nodes to select the desired number of seed nodes.")
        selected_remaining_seeds = random.sample(remaining_pool, remaining_seeds)
    else:
        selected_remaining_seeds = []

    # Combine all selected seeds
    seed_nodes = selected_product_seeds + selected_customer_seeds + selected_remaining_seeds
    return seed_nodes, selected_product_seeds, selected_customer_seeds, selected_remaining_seeds

# Parameters
num_seeds = 4  # Total number of seed nodes
desired_subgraph_size = 1000  # Target number of nodes
max_depth = 2  # Depth of BFS from each seed

# Select seed nodes
seed_nodes, seed_product_seeds, seed_customer_seeds, seed_remaining_seeds = select_seed_nodes(
    G, product_nodes, customer_nodes, num_seeds
)

print(f"Selected seed nodes (Products): {seed_product_seeds}")
print(f"Selected seed nodes (Customers): {seed_customer_seeds}")
if seed_remaining_seeds:
    print(f"Selected seed nodes (Remaining): {seed_remaining_seeds}")

# Step 2: Perform BFS from Each Seed and Track Connections
def bfs_expand(G, start_node, max_depth):
    """
    Performs BFS from a start_node up to max_depth.
    Returns the set of nodes visited during BFS.
    """
    queue = [(start_node, 0)]
    visited = set()
    while queue:
        current_node, depth = queue.pop(0)
        if current_node not in visited:
            visited.add(current_node)
            if depth < max_depth:
                neighbors = list(G.neighbors(current_node))
                for neighbor in neighbors:
                    if neighbor not in visited:
                        queue.append((neighbor, depth + 1))
    return visited

# Expand from each seed node using BFS and track connected nodes
seed_connections = {}
for seed in seed_nodes:
    connected_nodes = bfs_expand(G, seed, max_depth)
    seed_connections[seed] = connected_nodes
    print(f"Seed {seed} connected to {len(connected_nodes)} nodes.")

# Combine all visited nodes
visited_nodes = set()
for nodes in seed_connections.values():
    visited_nodes.update(nodes)

print(f"Total nodes collected after BFS: {len(visited_nodes)}")

# Step 3a: Assign Importance to Seed Nodes using PageRank
# Compute PageRank scores for all nodes in G
pagerank_scores = nx.pagerank(G)

# Extract PageRank scores for seed nodes
seed_pagerank = {seed: pagerank_scores.get(seed, 0) for seed in seed_nodes}

# Normalize seed PageRank scores
total_pr = sum(seed_pagerank.values())
if total_pr == 0:
    # Avoid division by zero
    seed_importance = {seed: 1 for seed in seed_nodes}
else:
    seed_importance = {seed: pr / total_pr for seed, pr in seed_pagerank.items()}

print("Seed Importance (PageRank):")
for seed, importance in seed_importance.items():
    print(f"Seed Node: {seed}, Importance: {importance:.4f}")

# Step 3b: Score Nodes in the Subgraph based on Seed Importance
def score_nodes_based_on_seeds(G, subgraph_nodes, seed_importance, max_depth):
    """
    Assigns a score to each node in subgraph_nodes based on the importance of seed nodes
    they are connected to within max_depth.
    """
    node_scores = {node: 0 for node in subgraph_nodes}

    for seed, importance in seed_importance.items():
        # Perform BFS from the seed up to max_depth
        queue = [(seed, 0)]
        visited = set()
        while queue:
            current_node, depth = queue.pop(0)
            if current_node in visited:
                continue
            visited.add(current_node)
            if current_node in node_scores:
                # Update the node's score based on seed importance and proximity
                # Closer nodes get higher scores
                proximity_score = (max_depth - depth + 1) / (max_depth + 1)  # Normalized proximity
                node_scores[current_node] += importance * proximity_score
            if depth < max_depth:
                neighbors = list(G.neighbors(current_node))
                for neighbor in neighbors:
                    if neighbor not in visited:
                        queue.append((neighbor, depth + 1))

    return node_scores

# Calculate node scores
node_scores = score_nodes_based_on_seeds(G, visited_nodes, seed_importance, max_depth)

# Step 3c: Prune the Subgraph to Desired Size based on Scores
def prune_subgraph(G, node_scores, desired_size):
    """
    Prunes the subgraph to the desired_size by selecting nodes with highest scores.
    """
    # Sort nodes by score in descending order
    sorted_nodes = sorted(node_scores.items(), key=lambda x: x[1], reverse=True)

    # Select top N nodes
    selected_nodes = [node for node, score in sorted_nodes[:desired_size]]

    # Construct the pruned subgraph
    pruned_subgraph = G.subgraph(selected_nodes).copy()

    return pruned_subgraph

# Check if pruning is needed
if len(visited_nodes) > desired_subgraph_size:
    print("Pruning the subgraph based on seed importance...")
    pruned_subgraph = prune_subgraph(G, node_scores, desired_subgraph_size)
    print(f"Pruned subgraph size: {len(pruned_subgraph.nodes())} nodes")
else:
    pruned_subgraph = G.subgraph(visited_nodes).copy()
    print(f"Subgraph size is within the desired limit: {len(pruned_subgraph.nodes())} nodes")

# Step 4: Simplify the subgraph and assign node types based on 'node_type' attribute
simplified_subgraph = nx.Graph()

for node, data in pruned_subgraph.nodes(data=True):
    node_type = data.get('node_type', '').lower()
    if node_type in ['product', 'customer']:
        simplified_subgraph.add_node(node, node_type=node_type)
    else:
        simplified_subgraph.add_node(node, node_type='unknown')

# Add edges with attributes
simplified_subgraph.add_edges_from(pruned_subgraph.edges(data=True))

# Update node types after potential modifications
node_types = nx.get_node_attributes(simplified_subgraph, 'node_type')

# Assign colors based on node type
color_map = []
for node in simplified_subgraph.nodes():
    if node_types[node] == 'product':
        color_map.append('#ff7f00')  # Orange color for products
    elif node_types[node] == 'customer':
        color_map.append('#33a02c')  # Green color for customers
    else:
        color_map.append('gray')     # Gray color for unknown

# Prepare positions for nodes using spring layout
pos = nx.spring_layout(simplified_subgraph, seed=42)

# Draw nodes
plt.figure(figsize=(14, 14))
nx.draw_networkx_nodes(
    simplified_subgraph,
    pos,
    node_color=color_map,
    node_size=100,
    alpha=0.8
)

# Draw edges
nx.draw_networkx_edges(
    simplified_subgraph,
    pos,
    edge_color='gray',
    alpha=0.5
)

# Optionally, draw labels for seed nodes
labels = {node: node for node in seed_nodes if node in simplified_subgraph}
nx.draw_networkx_labels(
    simplified_subgraph,
    pos,
    labels=labels,
    font_size=10,
    font_color='black',
    font_weight='bold'
)

plt.title("Pruned Subgraph Visualization with Node Types (Intelligent Pruning)")
plt.axis('off')  # Hide axes for a cleaner look

# Add legend for node colors
legend_elements = [
    Line2D([0], [0], marker='o', color='w', label='Product',
           markerfacecolor='#ff7f00', markersize=10),
    Line2D([0], [0], marker='o', color='w', label='Customer',
           markerfacecolor='#33a02c', markersize=10),
    Line2D([0], [0], marker='o', color='w', label='Unknown',
           markerfacecolor='gray', markersize=10)
]
plt.legend(handles=legend_elements, loc='best')

plt.show()

# Step 5: Analyze Pruning Per Seed
def analyze_pruning(seed_connections, pruned_subgraph_nodes):
    """
    Analyzes how much pruning occurred for each seed node.
    Returns a dictionary with seed nodes as keys and pruning statistics as values.
    """
    pruning_stats = {}
    for seed, connected_nodes in seed_connections.items():
        total_connected = len(connected_nodes)
        retained_connected = len(connected_nodes.intersection(pruned_subgraph_nodes))
        pruned_connected = total_connected - retained_connected
        retention_rate = retained_connected / total_connected if total_connected > 0 else 0
        pruning_stats[seed] = {
            'Total Connected Nodes': total_connected,
            'Retained Nodes': retained_connected,
            'Pruned Nodes': pruned_connected,
            'Retention Rate (%)': round(retention_rate * 100, 2)
        }
    return pruning_stats

# Get pruned subgraph nodes
pruned_nodes = set(pruned_subgraph.nodes())

# Analyze pruning
pruning_statistics = analyze_pruning(seed_connections, pruned_nodes)

# Display pruning statistics
print("\nPruning Statistics per Seed:")
for seed, stats in pruning_statistics.items():
    print(f"\nSeed Node: {seed}")
    print(f"  Total Connected Nodes: {stats['Total Connected Nodes']}")
    print(f"  Retained Nodes: {stats['Retained Nodes']}")
    print(f"  Pruned Nodes: {stats['Pruned Nodes']}")
    print(f"  Retention Rate: {stats['Retention Rate (%)']}%")
```
